# 장르의 평점 많은 순 정렬

```python
total = pd.read_csv('댓글합.csv', encoding='cp949')
total
```

```python
#장르가 '소설문학','오디오북스','자기계발','경제경영','인문교양','역사','어린이동화' 7개가 있다
df = total[['nickname', 'rating',  'genre']]
df
#장르마다 평점의 갯수를 세아리고
genre_df = df.groupby(['genre']).count().sort_values('nickname').reset_index().drop(['nickname'],axis=1)
genre_df
#세아린 평점의 갯수를 장르마다 비교해서, 평점이 많은 순으로 정렬
genre_df.sort_index(ascending=False)
```
