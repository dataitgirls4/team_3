스토리텔링과, 유저들이 어떻게 쓰고 있는지, 배우가 읽어주는 것에 대해서 더 피드백을 많이 한다?, 서포트를 할 여러가지의 근거를 만들어야함. 리뷰많이쓴것 ㅇㅇ, 리뷰 몇개 이상 해야할지 데이터를 봐야함. 나머지 사람들은 그 것을 보자

## 댓글 개수 8개 이상인 유저

```python
word_cnt = dict() 

for name in booklist['nickname']: 
    if name not in word_cnt.keys(): 
        word_cnt[name] = 1 
    else: 
        word_cnt[name] += 1
print(word_cnt)
```

```python
over_8_dict = {}
for key in word_cnt.keys() :
    if word_cnt.get(key) >= 8 :
        over_8_dict[key] = word_cnt.get(key)

len(over_8_dict)
```
