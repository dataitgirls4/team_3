```python
import pandas as pd
import numpy as np
```

```python
pod = pd.read_csv('댓글합.csv', encoding='cp949').drop('Unnamed: 0', axis = 1)
pod
```

```python
pod = pod[~pod['review'].isnull()]
pod = pod.reset_index().drop('index', axis=1)
pod
```

```python
def get_jaccard_sim(str1, str2): 
    a = set(str1.split()) 
    b = set(str2.split())
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))
```

```python
a=[]
for i in range(15480):
    str1 = pod['review'][i]
    for j in range(i+1, 15480):
        str2 = pod['review'][j]
        a.append(get_jaccard_sim(str1, str2))
```
