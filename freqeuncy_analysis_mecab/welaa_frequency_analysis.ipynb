```python
w = w[~w['review'].isnull()]
w = w.reset_index().drop('index', axis=1)
w
```

```python
def get_jaccard_sim(str1, str2): 
    a = set(str1.split()) 
    b = set(str2.split())
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))
```

```python
k=[]
for i in range(4037):
    str1 = w['review'][i]
    for j in range(i+1, 4037):
        str2 = w['review'][j]
        if  len(set(str1.split())) +len(set(str2.split())) - len(set(str1.split()).intersection(set(str2.split()))) == 0:
            pass
        else:
            k.append(get_jaccard_sim(str1, str2))
```
