```python
import pandas as pd
import numpy as np
```

```python
n = pd.read_csv('naver_imsi_review.csv', encoding='utf-8').drop('Unnamed: 0', axis = 1)
n
```

```python
n = n[~n['review'].isnull()]
n = n.reset_index().drop('index', axis=1)
n
```

```python
def get_jaccard_sim(str1, str2): 
    a = set(str1.split()) 
    b = set(str2.split())
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))
```

```python
a=[]
for i in range(1305):
    str1 = n['review'][i]
    for j in range(i+1, 1305):
        str2 = n['review'][j]
        a.append(get_jaccard_sim(str1, str2))
```

```python
np.mean(a)
```
